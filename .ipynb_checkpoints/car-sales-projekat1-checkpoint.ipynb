```python
# kreiranje matrice sa procentima
def calculate_percentages(df, input_column, target_column):
    # Grupišemo podatke po input koloni i računamo procenat za svaku vrednost target kolone
    grouped = df.groupby(input_column)[target_column].value_counts(normalize=True).unstack()
    
    # Množimo sa 100 da dobijemo procente i zaokružujemo na 2 decimale
    percentages = (grouped * 100).round(2)
    
    # Dodajemo kolonu sa ukupnim brojem instanci za svaku vrednost input kolone
    percentages['Total Count'] = df[input_column].value_counts()
    
    return percentages
```

```python
# funkcija za kreiranje grafika gde su na y osi procenti raspodele
def create_percentage_chart(df, input_column, target_column):
    # Računamo procente
    percentages = df.groupby(input_column)[target_column].value_counts(normalize=True).unstack() * 100

    # Kreiramo grafikon
    ax = percentages.plot(kind='bar', stacked=True, figsize=(12, 6))

    # Podešavamo naslov i labele
    plt.title(f'Raspodela {target_column} za različite vrednosti {input_column}', fontsize=16)
    plt.xlabel(input_column, fontsize=12)
    plt.ylabel('Procenat', fontsize=12)

    # Rotiramo labele na x-osi ako ih ima mnogo
    plt.xticks(rotation=45, ha='right', fontsize=12)

    # Dodajemo legendu
    plt.legend(title=target_column, bbox_to_anchor=(1.05, 1), loc='upper left')

    # Dodajemo procente na stubićima
    for c in ax.containers:
        ax.bar_label(c, fmt='%.1f%%', label_type='center')

    # Podešavamo layout da se sve vidi
    plt.tight_layout()

    # Prikazujemo grafikon
    plt.show()
```

Provericemo da li izlazna varijabla zavisi od preostalih atributa i preko statistickih testova.

```python
#SPIRMANOV TEST
numericki = train.select_dtypes(include='number')
from scipy.stats import spearmanr
results = []
y = train['IsBadBuy']
for col in numericki.columns:
    correlation_coefficient, p_value = spearmanr(train[col], y)
    results.append((col, correlation_coefficient, p_value))
    if p_value < 0.05:
        print('Postoji zavisnot izmedju x i y za',col)
    else:
        print('Ne postoji statisticka znacajna zavisnost za',col)
```

```python
#HI KVADRAT
from scipy.stats import chi2_contingency

y = train['IsBadBuy']
object_columns = [col for col in train.columns if train[col].dtype == 'object']
def chi2_test(col1,target):
    contingency_table = pd.crosstab(train[col1],target)
    chi2, p, _, _ = chi2_contingency(contingency_table)
    return p

for i in range(len(object_columns)):
    col1 = object_columns[i]
    p_value = chi2_test(col1,y)
    results.append((col1, 'y', p_value))
    if p_value < 0.05:
        print('Postoji zavisnot izmedju x i y za',object_columns[i])
    else:
        print('Ne postoji statisticka znacajna zavisnost za', object_columns[i])
```
